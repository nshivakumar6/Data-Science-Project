```
import pandas as pd
```

```
df = pd.read_csv('new_avgtempatl.csv')
df.head()
```

```
from sklearn.model_selection import train_test_split
X = df.drop(columns = ['Year', 'Winter', 'Summer', 'Fall', 'Spring', 'Jan', 'Unnamed: 0', 'Annual'])
y = df['Jan']

# train test splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle = True)
```

```
from sklearn.linear_model import LinearRegression
```

```
linmodel = LinearRegression()
linmodel.fit(X_train,y_train)
```

```
predictions = linmodel.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(mean_squared_error(y_test, predictions))
print(mean_absolute_error(y_test, predictions))
```

```
from sklearn.neural_network import MLPRegressor

nn = MLPRegressor(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(5, 5), random_state=5)

nn.fit(X_train, y_train)

print('Training accuracy: ', nn.score(X_train, y_train))
print('Testing accuracy: ', nn.score(X_test, y_test))
```
