```
import pandas as pd
```

```
df = pd.read_csv('clean_avgtempatl.csv')
```

```
df.head()
```

```
df = df.drop("Unnamed: 0", axis=1)
```

```
df["Winter"] = round(((df["Dec"] + df["Jan"] + df["Feb"])/3),1)
df["Spring"] = round(((df["Mar"] + df["Apr"] + df["May"])/3), 1)
df["Summer"] = round(((df["Jun"] + df["Jul"] + df["Aug"])/3), 1)
df["Fall"] = round(((df["Sep"] + df["Oct"] + df["Nov"])/3), 1)
```

```
df.head()
```

```
df.to_csv("new_avgtempatl.csv")
```
