```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('avgtempatl.csv')
```

```
df = df.drop(labels=range(0,2), axis=0)
df = df.drop(labels=range(43,53), axis=0)
df = df.drop(labels=range(62,71), axis=0)
df = df.drop("Unnamed: 14", axis=1)
df
```

```
df = df.rename(columns={"10/20/22, 2:13 PM":"Year", "Unnamed: 1":"Jan", "Unnamed: 2":"Feb", 
"Unnamed: 3":"Mar", "Unnamed: 4":"Apr", "Unnamed: 5":"May", "Unnamed: 6":"Jun", 
"Unnamed: 7":"Jul", "about:blank":"Aug", "Unnamed: 9":"Sep", "Unnamed: 10":"Oct",
"Unnamed: 11":"Nov", "Unnamed: 12":"Dec", "Unnamed: 13":"Annual"})

monthslist = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Annual"]
for month in monthslist:
    df[month] = df[month].astype(float)
```

```
for month in monthslist:
    print("The mean for {} is ".format(month))
    print(str(round(sum(df[month])/len(df[month]),2)) + "\n")
```

```
plt.hist(df["Annual"], bins=20);
```

```
plt.bar(df["Year"], df["Feb"], tick_label = "")
plt.xlabel("Year, 1972-2021")
plt.ylabel("Average Temperature in February")
plt.ylim(30,70);
```

```
plt.bar(df["Year"], df["Jul"], tick_label = "")
plt.xlabel("Year, 1972-2021")
plt.ylabel("Average Temperature in July")
plt.ylim(70,90);
```

```
plt.bar(df["Year"], df["Annual"], tick_label = "")
plt.xlabel("Year, 1972-2021")
plt.ylabel("Average Temperature Annually")
plt.ylim(55,70);
```

```
df.to_csv("clean_avgtempatl.csv")
```
